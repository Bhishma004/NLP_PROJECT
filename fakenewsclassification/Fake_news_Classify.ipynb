```python
import tensorflow as tf
tf.__version__
```

```python
from tensorflow.keras.layers import Embedding
from tensorflow.keras.preprocessing.sequence import pad_sequences
from tensorflow.keras.models import Sequential
from tensorflow.keras.preprocessing.text import one_hot
from tensorflow.keras.layers import LSTM
from tensorflow.keras.layers import Dense
```

```python
voc_size = 5000
```

```python
messages = x.copy()
messages['title'][6]
```

```python
len(messages)
```

```python
import nltk
import re
from nltk.corpus import stopwords
```

```python
nltk.download('stopwords')
```

```python
from nltk.stem.porter import PorterStemmer
ps = PorterStemmer()
```

```python
corpus = []
for i in range(0, len(messages)):
    review = re.sub('[^a-zA-Z]', ' ', messages['title'][i])
    review = review.lower()
    review = review.split()
    
    review = [ps.stem(word) for word in review if not word in stopwords.words('english')]
    review = ' '.join(review)
    corpus.append(review)
```

```python
corpus[1]
```

```python
corpus
```

```python
onehot_repr = [one_hot(words,voc_size) for words in corpus]
onehot_repr
```

```python
onehot_repr[1]
```

```python
sent_length = 20
embedded_docs=pad_sequences(onehot_repr,padding='post',maxlen=sent_length)
print(embedded_docs)
```

```python
## Creating model
embedding_vector_features=40 ##features representation

#A Sequential model is appropriate for a plain stack of layers where each layer has exactly one
#input tensor and one output tensor.
model=Sequential()

# Embedding layer enables us to convert each word into a fixed length vector of defined size
model.add(Embedding(voc_size,embedding_vector_features,input_length=sent_length))
```

```python

model.add(LSTM(100))
```

```python
#Dense layer is the regular deeply connected neural network layer.
model.add(Dense(1,activation='sigmoid'))

model.compile(loss='binary_crossentropy',optimizer='adam',metrics=['accuracy'])
print(model.summary())
```

```python
import numpy as np
X_final=np.array(embedded_docs)
y_final=np.array(y)
X_final.shape,y_final.shape
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X_final, y_final, test_size=0.33, random_state=42)
```

```python
model.fit(X_train,y_train,validation_data=(X_test,y_test),epochs=10,batch_size=64)
```

```python
from sklearn.metrics import confusion_matrix


y_pred=model.predict(X_test)


y_pred=np.where(y_pred > 0.6, 1,0) ##AUC ROC Curve

confusion_matrix(y_test,y_pred)
```

```python
from sklearn.metrics import accuracy_score
accuracy_score(y_test,y_pred)
```

```python
from sklearn.metrics import classification_report
print(classification_report(y_test,y_pred))
```
